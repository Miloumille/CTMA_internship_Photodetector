```python
import PySimpleGUI as sg
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Load the image
Img = "recent_correct_data/TestPic_3_XRGB8888_Br2_Cst2.jpg"
imageInput = cv2.imread(Img)

# Get the image dimensions
height, width, _ = imageInput.shape

# Predefined position values for vertical and horizontal lines (grid positions)
vertical_positions = [40, 83, 128, 175, 225, 270, 310, 350, 400, 445, 490, 530, 580]
horizontal_positions = [0, 60, 120, 175, 235, 290, 345, 405, 465]

# Function to draw the grid based on the position vectors
def draw_grid(image, vertical_positions, horizontal_positions):
    # Draw vertical grid lines
    for x in vertical_positions:
        cv2.line(image, (x, 0), (x, height), (255, 255, 255), 1)  # Vertical lines in white
    # Draw horizontal grid lines
    for y in horizontal_positions:
        cv2.line(image, (0, y), (width, y), (255, 255, 255), 1)  # Horizontal lines in white

    # Add labels to each grid cell
    grid_height = len(horizontal_positions) - 1
    grid_width = len(vertical_positions) - 1
    cell_dict = {}
    
    for i in range(grid_height):
        for j in range(grid_width):
            label = chr(65 + i) + str(j + 1)  # Create names like A1, A2, ...
            cell_dict[label] = {
                'x_start': vertical_positions[j],
                'x_end': vertical_positions[j + 1],
                'y_start': horizontal_positions[i],
                'y_end': horizontal_positions[i + 1]
            }
            # Draw label on the image
            cv2.putText(image, label, (vertical_positions[j] + 5, horizontal_positions[i] + 30),
                        cv2.FONT_HERSHEY_SIMPLEX, 0.5, (255, 255, 255), 1)

    return cell_dict

# Draw the grid and get cell coordinates
cell_dict = draw_grid(imageInput, vertical_positions, horizontal_positions)

# Select a cell by name (e.g., 'A1')
cell_name = 'D7'  # Change this to any cell name you want to focus on

# Get the coordinates from the dictionary
if cell_name in cell_dict:
    cell_coords = cell_dict[cell_name]
    cell_x_start = cell_coords['x_start']
    cell_x_end = cell_coords['x_end']
    cell_y_start = cell_coords['y_start']
    cell_y_end = cell_coords['y_end']

    # Crop the sub-image for the selected cell
    cell_image = imageInput[cell_y_start:cell_y_end, cell_x_start:cell_x_end]

    # Plotting the histogram for the selected cell in grayscale
    gray_cell_image = cv2.cvtColor(cell_image, cv2.COLOR_BGR2GRAY)

    # Plotting the histogram
    plt.figure(figsize=(8, 4))
    plt.hist(gray_cell_image.ravel(), bins=256, range=[0, 256], color='gray')
    plt.title(f'Histogram of Cell {cell_name}')
    plt.xlabel('Pixel Intensity')
    plt.ylabel('Frequency')
    plt.grid()
    plt.show()
else:
    print(f"Cell '{cell_name}' not found in the grid.")

# Convert the image back to a format suitable for PySimpleGUI
imageBytes = cv2.imencode('.png', imageInput)[1].tobytes()

# Create a PySimpleGUI window to display the image with the grid
layout = [[sg.Image(data=imageBytes)]]
window = sg.Window('Image with Grid Overlay', layout, finalize=True)

# Event loop to keep the window open
while True:
    event, values = window.read()
    if event == sg.WIN_CLOSED:
        break

window.close()
```

```python
import cv2
import numpy as np

# Load image
Img = "recent_correct_data/TestPic_3_XRGB8888_Br2_Cst2.jpg"
imageInput = cv2.imread(Img)

# Convert to LAB color space
lab = cv2.cvtColor(imageInput, cv2.COLOR_BGR2LAB)
l, a, b = cv2.split(lab)

# Apply CLAHE to L-channel for adaptive equalization
clahe = cv2.createCLAHE(clipLimit=5.0, tileGridSize=(20, 20))
cl = clahe.apply(l)

# Merge and convert back to BGR
limg = cv2.merge((cl, a, b))
balanced_image = cv2.cvtColor(limg, cv2.COLOR_LAB2BGR)

plt.imshow(cv2.cvtColor(balanced_image, cv2.COLOR_BGR2RGB))
plt.axis('off')  # Hide axes for a cleaner look
plt.show()
```

```python
b, g, r = cv2.split(balanced_image)

# Define a blank channel for isolating colors
blank_channel = np.zeros_like(b)

# Create images with only one color channel visible at a time
red_only = cv2.merge([blank_channel, blank_channel, r])    # Red channel
green_only = cv2.merge([blank_channel, g, blank_channel])  # Green channel
blue_only = cv2.merge([b, blank_channel, blank_channel])   # Blue channel

# Plot the images
fig, axs = plt.subplots(1, 3, figsize=(15, 5))
axs[0].imshow(cv2.cvtColor(blue_only, cv2.COLOR_BGR2RGB))
axs[0].set_title("Blue Channel Only")
axs[0].axis('off')

axs[1].imshow(cv2.cvtColor(green_only, cv2.COLOR_BGR2RGB))
axs[1].set_title("Green Channel Only")
axs[1].axis('off')

axs[2].imshow(cv2.cvtColor(red_only, cv2.COLOR_BGR2RGB))
axs[2].set_title("Red Channel Only")
axs[2].axis('off')

plt.show()
```
